```python
from automata.fa.dfa import DFA

from visual_automata.fa.dfa import VisualDFA
```

```python
id_A = DFA(
    states={'start', 's0'},
    input_symbols={"123…", "ABC…", "abc…"},
    transitions={
        'start': {'ABC…': 's0', 'abc…': 's0'},

        's0': {'123…': 's0', 'ABC…': 's0', 'abc…': 's0'},
    },
    initial_state='start',
    final_states={'s0'},
    allow_partial=True
)
```

```python
visual_id_A = VisualDFA(id_A)
visual_id_A.show_diagram()
```

```python
num_B = DFA(
    states={'start', 's0'},
    input_symbols={'0…9'},
    transitions={
        'start': {'0…9': 's0'},

        's0': {'0…9': 's0'},
    },
    initial_state='start',
    final_states={'s0'},
    allow_partial=True
)
```

```python
visual_num_B = VisualDFA(num_B)
visual_num_B.show_diagram()
```

```python
semicolon_C = DFA(
    states={'start', 's0', 's1'},
    input_symbols={';'},
    transitions={
        'start': {';': 's0'},
        's0': {';': 's1'},
        's1': {},
    },
    initial_state='start',
    final_states={'s0'},
    allow_partial=True
)
```

```python
visual_semicolon_C = VisualDFA(semicolon_C)
visual_semicolon_C.show_diagram()
```

```python
and_D = DFA(
    states={'start', 's0', 's1', 's2'},
    input_symbols={'&'},
    transitions={
        'start': {'&': 's0'},
        's0': {'&': 's1'},
        's1': {'&': 's2'},
        's2': {}
    },
    initial_state='start',
    final_states={'s1'},
    allow_partial=True
)
```

```python
visual_and_D = VisualDFA(and_D)
visual_and_D.show_diagram()
```
